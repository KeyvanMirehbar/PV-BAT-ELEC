```python
import pandas as pd
import matplotlib.pyplot as plt
import calendar
import os
import tempfile
import shutil

# Helper functions to modularize the code
def load_pv_data(pv_choice):
    pv_size_options = {
        1: r'C:\Users\keyvan.mirehbar\Desktop\PV data\10kW_PV_data.xlsx',
        2: r'C:\Users\keyvan.mirehbar\Desktop\PV data\20kW_PV_data.xlsx',
        3: r'C:\Users\keyvan.mirehbar\Desktop\PV data\50kW_PV_data.xlsx',
        4: r'C:\Users\keyvan.mirehbar\Desktop\PV data\100kW_PV_data.xlsx'
    }
    
    if pv_choice in pv_size_options:
        file_path = pv_size_options[pv_choice]
        df = pd.read_excel(file_path)
        df_cleaned = df.dropna(how='all')  # Remove rows with all NaN values
        df_cleaned['Hour'] = df_cleaned['Hour'].str.split(' - ').str[0].astype(int)
        df_cleaned.set_index('Hour', inplace=True)
        return df_cleaned
    else:
        raise ValueError("Invalid selection for PV plant size.")

def load_price_data(price_file_path):
    price_df = pd.read_excel(price_file_path)
    if price_df['Price'].isnull().any():
        raise ValueError("Electricity price data contains missing values.")
    return price_df

def calculate_hydrogen_production(electrolyzer_power, hour, degradation_factor, hydrogen_production_rate):
    return electrolyzer_power * hydrogen_production_rate * (1 - degradation_factor)**hour

def simulate_battery_system(df_cleaned, price_df, electrolyzer_power, battery_capacity, initial_soc, hydrogen_production_rate, simulation_hours, degradation_factor, grid_outage_hours, battery_degradation_rate):
    soc = initial_soc
    soc_values = []
    grid_usage = []
    hydrogen_green = 0  # Total green hydrogen
    hydrogen_yellow = 0  # Total yellow hydrogen
    previous_soc = soc
    cycle_count = 0
    
    for hour in range(simulation_hours):
        pv_generation = df_cleaned.iloc[hour % 24][month]  # Access specific value for the current hour and month
        electricity_price_hour = float(price_df.loc[hour % 24, 'Price'])
        grid_power = float(0)  # Default value for grid power usage
        
        # Case 1: PV generation exceeds electrolyzer demand
        if pv_generation > electrolyzer_power:
            excess_power = pv_generation - electrolyzer_power
            soc = min(soc + excess_power, battery_capacity)
            # Green hydrogen production
            hydrogen_produced_m3 = calculate_hydrogen_production(electrolyzer_power, hour, degradation_factor, hydrogen_production_rate)
            hydrogen_green += hydrogen_produced_m3
        
        # Case 2: PV generation is less than electrolyzer demand
        else:
            power_deficit = electrolyzer_power - pv_generation
            if soc > 0:
                soc = max(soc - power_deficit, 0)
                # Green hydrogen from battery discharge
                hydrogen_produced_m3 = calculate_hydrogen_production(electrolyzer_power, hour, degradation_factor, hydrogen_production_rate)
                hydrogen_green += hydrogen_produced_m3
            else:
                if hour not in grid_outage_hours:
                    grid_power = float(power_deficit)
                    # Yellow hydrogen from grid electricity
                    hydrogen_produced_m3 = calculate_hydrogen_production(electrolyzer_power, hour, degradation_factor, hydrogen_production_rate)
                    hydrogen_yellow += hydrogen_produced_m3
        
        # Check if a full charge-discharge cycle occurred
        if previous_soc < initial_soc and soc == battery_capacity:
            cycle_count += 1
            battery_capacity *= (1 - battery_degradation_rate)  # Apply degradation per cycle
            print(f"Cycle {cycle_count}: Battery capacity degraded to {battery_capacity:.2f} kWh")
        
        previous_soc = soc
        
        soc_values.append(soc)
        grid_usage.append(grid_power)
    
    return soc_values, grid_usage, hydrogen_green, hydrogen_yellow

def plot_pie_chart(green_hydrogen, yellow_hydrogen, month, output_dir):
    labels = ['Green Hydrogen', 'Yellow Hydrogen']
    sizes = [green_hydrogen, yellow_hydrogen]
    colors = ['green', 'yellow']
    
    fig, ax = plt.subplots()
    ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'black'})
    ax.set_title(f'Hydrogen Production for {month}')
    plt.savefig(os.path.join(output_dir, f'Hydrogen_Production_{month}.png'))  # Save the pie chart
    plt.show()

def zip_results(output_dir, zip_filename):
    """Compresses the results into a zip file."""
    shutil.make_archive(zip_filename, 'zip', output_dir)
    print(f"All results zipped successfully into {zip_filename}.zip")

# Main code execution
output_dir = tempfile.mkdtemp()  # Create a temporary directory for saving all files
zip_filename = r'C:\Users\keyvan.mirehbar\Desktop\PV_data_results'  # Define where to save the zip file

try:
    pv_choice = int(input("Select the size of the PV plant (1. 10kW, 2. 20kW, 3. 50kW, 4. 100kW): "))
    price_file_path = r'C:\Users\keyvan.mirehbar\Desktop\PV data\Price 2.xlsx'  # Assuming this is already set

    df_cleaned = load_pv_data(pv_choice)
    price_df = load_price_data(price_file_path)

    electrolyzer_power = float(input("Enter the power consumption of the electrolyzer (in kW): "))
    battery_capacity = float(input("Enter the battery capacity (in kWh): "))
    initial_soc = float(input("Enter the initial State of Charge (SOC) of the battery in kWh): "))
    hydrogen_production_rate = float(input("Enter the hydrogen production rate in mÂ³ per kW per hour: "))
    degradation_factor = float(input("Enter the electrolyzer degradation factor per hour (e.g., 0.001 for 0.1%): "))
    battery_degradation_rate = float(input("Enter the battery degradation rate per cycle (e.g., 0.001 for 0.1%): "))

    grid_outage_hours = [18, 19, 20]  # Example of peak-hour grid outages

    months = df_cleaned.columns.tolist()

    for month in months:
        # Get the number of days in the current month
        days_in_month = calendar.monthrange(2024, months.index(month) + 1)[1]  # Assuming year 2024
        simulation_hours = days_in_month * 24  # Total hours in the month
        
        soc_values, grid_usage, hydrogen_green, hydrogen_yellow = simulate_battery_system(
            df_cleaned, price_df, electrolyzer_power, battery_capacity, initial_soc, hydrogen_production_rate,
            simulation_hours, degradation_factor, grid_outage_hours, battery_degradation_rate
        )
        
        # Plot and save 2D pie chart for each month in the temp directory
        plot_pie_chart(hydrogen_green, hydrogen_yellow, month, output_dir)

    # Zip all the results
    zip_results(output_dir, zip_filename)

finally:
    # Cleanup: Remove the temporary directory after zipping the files
    shutil.rmtree(output_dir)
    print(f"Temporary directory {output_dir} cleaned up.")
```
